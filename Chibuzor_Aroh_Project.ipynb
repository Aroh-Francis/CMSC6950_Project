```python
import pandas as pd
import numpy as np
```

### Load the dataset

```python

df = pd.read_csv('en_climate_monthly_NL_8403506_1942-2012_P1M.csv')
```

### Clean the Data (Data Preprocessing)

```python
# Handle Missing Values 

flaged_columns = [col for col in df.columns if 'Flag' in col]
is_df_cleaned = df.drop(columns=flaged_columns)
is_core_cols = [
            'Mean Max Temp (°C)', 'Mean Min Temp (°C)', 
            'Mean Temp (°C)','Extr Max Temp (°C)', 
            'Extr Min Temp (°C)', 'Total Rain (mm)',
            'Total Snow (cm)', 'Total Precip (mm)'
            ]

core_data = is_df_cleaned[is_core_cols]
rows_with_missing = core_data.isnull().any(axis=1)
missing_data = is_df_cleaned[rows_with_missing]

print(f"\n We Found {len(missing_data)} rows with missing core data and they will be dropped.")
is_df_cleaned = is_df_cleaned.dropna(subset=is_core_cols)


# Adjust the Date and Time format and save file to a new file

is_df_cleaned['Date/Time'] = pd.to_datetime(is_df_cleaned['Date/Time'])
is_df_cleaned.to_csv('cleaned_climate_file.csv', index=False)
print("\n The data cleaning process is complete. Data saved to 'cleaned_climate_file.csv'")

print(f"\n Original data shape: {df.shape}")
print(f"\n Cleaned data shape: {is_df_cleaned.shape}")
```
